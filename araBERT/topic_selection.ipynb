```python
# Filter out rows where 'Topic_new' equals 'noise'
df = df[df['Topic_new'] != 'noise']
```

```python

# Step 1: Group df by 'Comment_ID' and aggregate the 'Topic_new' values into a list
topics_list = df.groupby('Comment_ID')['Topic_new'].apply(list).reset_index(name='Topics')

# Step 2: Merge this list into df_big based on the Comment_ID
df_big = df_big.merge(topics_list, on='Comment_ID', how='left')

```

```python
df_big
```

```python
df_big.dropna(subset=["Topics"], inplace=True)
```

```python
df_big["Topics"] = df_big["Topics"].apply(
    lambda topic_list: list(dict.fromkeys(topic_list)) if isinstance(topic_list, list) else topic_list
)
```

```python
df_big
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd


# Initialize the binarizer and transform the Topics column
mlb = MultiLabelBinarizer()
topics_binary = pd.DataFrame(mlb.fit_transform(df_big["Topics"]),
                             columns=mlb.classes_,
                             index=df_big.index)

# Concatenate the binary columns to your original DataFrame
df_result = pd.concat([df_big, topics_binary], axis=1)

# Now, every comment is in a row and each topic is a separate column
# with a 1 indicating the comment includes that topic, and 0 otherwise.
df_result
```

```python
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta


# Function to subtract a duration from todays date 
def subtract_duration(ref_date, duration_str):
    if isinstance(duration_str, str):
        if duration_str.lower() == 'unknown':
            return None
        try:
            value = int(duration_str[:-1])
        except ValueError:
            return None
        unit = duration_str[-1]
        if unit == 'y':
            return ref_date - relativedelta(years=value)
        elif unit == 'm':  # here 'm' means months
            return ref_date - relativedelta(months=value)
        elif unit == 'w':
            return ref_date - timedelta(weeks=value)
        elif unit == 'd':
            return ref_date - timedelta(days=value)
        elif unit == 'h':
            return ref_date - timedelta(hours=value)
    return None

ref_date = datetime(2025, 3, 18, 12, 0)


df_result['computed_date'] = df_result['Time_Elapsed'].apply(lambda x: subtract_duration(ref_date, x))
```
